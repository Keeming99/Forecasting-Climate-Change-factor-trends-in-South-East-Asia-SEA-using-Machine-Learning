### ii. Data Cleaning and Preparation for SEA countries data

In both datasets, unnecessary columns were dropped, and remaining columns were renamed. For temperature change dataset, important columns like Country, ISO3, and temperature change from year 1961 to 2022 were maintained. For CO2 emissions dataset, important columns such as Entity, Code, and CO2 emissions columns from year 1949 to 2022 were maintained. In this step, temperature change dataset were reshaped from wide to long dataset according to country, using the melt function. The year data for temperature change dataset are prefix by ‘F’ e.g., F1961, F1962. Data processing will be performed to remove the ‘F’ character for each year and cast the year value to integer. Both datasets column name are renamed for better understanding.

```python
# Data Cleaning and Preparation for SEA countries

# Temperature Dataset
# Dropping unnecessary columns
temp_cols_to_drop = ['ObjectId', 'ISO2', 'Indicator', 'Unit', 'Source', 'CTS_Code', 'CTS_Name', 'CTS_Full_Descriptor']
temp_df_sea_cleaned = temp_df_sea.drop(columns=temp_cols_to_drop)

# Renaming 'ISO3' to 'Country_Code'
temp_df_sea_cleaned.rename(columns={'ISO3': 'Country_Code'}, inplace=True)

# Reshaping the dataset from wide to long format
temp_df_sea_cleaned = temp_df_sea_cleaned.melt(id_vars=['Country', 'Country_Code'],
                                            var_name='Year', value_name='Temperature_Change')
# Convert year from string to integer
temp_df_sea_cleaned['Year'] = temp_df_sea_cleaned['Year'].str[1:].astype(int)

temp_df_sea_cleaned.head()
```

```python
# CO2 Emissions Dataset
# Investigating the CO2 emissions dataset for SEA countries to identify duplicate or incorrectly labeled year columns
co2_df_sea.columns.tolist()

co2_df_sea_cleaned = co2_df_sea.copy()

# Renaming 'Country Name' to 'Country'
co2_df_sea_cleaned.rename(columns={'Entity': 'Country'}, inplace=True)

# Renaming 'country_code' column for alignment
co2_df_sea_cleaned.rename(columns={'Code': 'Country_Code'}, inplace=True)

# Renaming 'Annual CO₂ emissions (per capita)' to 'CO2_Per_Capita'
co2_df_sea_cleaned.rename(columns={'Annual CO₂ emissions (per capita)': 'CO2_Per_Capita'}, inplace=True)

co2_df_sea_cleaned.groupby(['Country_Code', 'Year'])['CO2_Per_Capita'].mean().reset_index()

# Display the first few rows of each cleaned and reshaped dataset
co2_df_sea_cleaned.head()
```

### iii. Null Value Imputation

For the current time series dataset, forward fill and backward fill were used to impute null values to minimise bias. Null values were completely removed after the imputation.

```python
# Imputation of Missing Data
# Merging the datasets on 'Country_Code' and 'Year'
combined_sea_df = pd.merge(temp_df_sea_cleaned, co2_df_sea_cleaned, on=['Country_Code', 'Year'])

# Remove and rename columns
combined_sea_df = combined_sea_df.drop(columns=['Country_x'])
combined_sea_df.rename(columns={'Country_y': 'Country'}, inplace=True)

# Checking for missing values
missing_values_before = combined_sea_df.isnull().sum()

combined_sea_df['Temperature_Change'] = combined_sea_df.groupby('Country')['Temperature_Change'].apply(lambda x: x.ffill().bfill())
combined_sea_df['CO2_Per_Capita'] = combined_sea_df.groupby('Country')['CO2_Per_Capita'].apply(lambda x: x.ffill().bfill())


# Rechecking for missing values after imputation
missing_values_after = combined_sea_df.isnull().sum()
# Display the first few rows of the merged dataset and the missing values status after imputation
combined_sea_df.head(), missing_values_before, missing_values_after
```

### iv. Final Dataset

For the final dataset, columns were rearranged using reindex.

```python
# Now everything looks good! year 1961 - 2022
# creating a list of columns
swap_list = ["Country","Country_Code","Year","Temperature_Change","CO2_Per_Capita"]

# Swapping the columns
combined_sea_df = combined_sea_df.reindex(columns=swap_list)

combined_sea_df
```

## 4.&nbsp;Data Modeling


### i. Stationary Data Transformation

Time series models requires data to be stationary. Non-stationary data exhibits varying mean, variance, or covariance over time, while stationary data lacks trends, seasonality, cyclical, or irregular patterns. The Augmented Dickey-Fuller (ADF) Test is used to evaluate data stationarity. If the p-value from the ADF Test exceeds the significance level (usually 0.05), it indicates a failure to reject the null hypothesis (H0) that the series is non-stationary. Conversely, a p-value lower than the significance level rejects the null hypothesis, suggesting stationarity.

```python
from statsmodels.tsa.vector_ar.var_model import VAR
from statsmodels.tsa.stattools import adfuller, pacf
from statsmodels.graphics.tsaplots import plot_pacf
from sklearn.metrics import mean_squared_error
from scipy.stats import pearsonr
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
```

```python
data = combined_sea_df.copy()

# Start modelling with data
# Preparing the data for the VAR model
# Find the mean for each year - indicating regional mean
combined_data = data.groupby("Year")[["Temperature_Change", "CO2_Per_Capita"]].mean()

# VAR is sensitive to data stationary, we will test it using Augmented Dickey-Fuller
def test_stationarity(data):
    adf_result = adfuller(data)
    print(data.name)
    print("------------------------------")
    print("ADF Statistic:", adf_result[0])
    print("p-value:", adf_result[1])
    print("Critical Values:")
    for key, value in adf_result[4].items():
        print("\t", key, ":", value)
    print("\n")

test_stationarity(combined_data["Temperature_Change"])
test_stationarity(combined_data["CO2_Per_Capita"])
```

The fact that the p-value is higher than the significant value of 0.05 and the ADF Stats is higher than all the critical values, we can clearly reject the null hypothesis and conclude that the dataset is not stationary. Futher difference is needed to make the data stationary

```python
def transform_data(data):
  # Return the differenced data
  return data.diff()

combined_data_diff = combined_data.copy()
combined_data_diff["Temperature_Change"] = transform_data(combined_data["Temperature_Change"])
test_stationarity(combined_data_diff["Temperature_Change"].dropna())

combined_data_diff["CO2_Per_Capita"] = transform_data(combined_data["CO2_Per_Capita"])
test_stationarity(combined_data_diff["CO2_Per_Capita"].dropna())

combined_data_diff.dropna(inplace = True)
```

The p-value is significant lower than the significant value of 0.05 and the ADF Stats is significant lower than all the critical values after data differencing, we can clearly accept the null hypothesis and conclude that the dataset is stationary. We can say data after differencing is working.

Split the difference dataset into train and test set. The test set is will be the last 5 years of the dataset. The training set is the rest of the data excluding the test set.

```python
df = combined_data_diff.copy()

# We set 5 observation (for prediction)
test_obs = 5
train = df[:-test_obs]
test = df[-test_obs:]
```

### ii. Vector AutoRegressive (VAR)

The Vector Auto Regressive (VAR) model stands out as a multivariate forecasting technique designed for situations where multiple time series mutually affect one another. Unlike other autoregressive models, VAR models factor in bidirectional relationships between these series.

The lag order (P) model is evaluated across a range from 1 to 10 by employing the Akaike Information Criterion (AIC) to assess performance.

```python
# Search through lag order from 1 to 10 and choose the best lag order
for i in range(1,11):
    var_model = VAR(train)
    results = var_model.fit(i)
    print('Order =', i)
    print('AIC: ', results.aic)
    print()
```

We can clearly see that the AIC increases after lag order 2 showing model of order 2 is suitable for us. The increases in AIC indicates that the complex model is being punished. We will use lag order 2 to fit the VAR model.

```python
# Fit the model according the significant Order of 2
result = var_model.fit(2)
result.summary()
```

We will then use the trained VAR model to forecast the next 5 years of temperature change and CO2 per capita change.

```python
# Create prediction - get the last 2 train value following the lag order number of previous observations
lagged_values = train.values[-2:]
var_pred = result.forecast(y=lagged_values, steps=test_obs)

# Create the Dataframe for the forecasted value
idx = ['2018','2019','2020','2021','2022']
df_var_forecast=pd.DataFrame(data=var_pred, index=idx, columns=['Temperature_Change', 'CO2_Per_Capita'])
```

Since we had already done differencing on the train set, we will then reverse differencing on the forecasted value to get the actual predicted value.

### iii. Autoregressive Integrated Moving Average (ARIMA)

The Autoregressive Integrated Moving Average (ARIMA) model is a statistical approach that employs time series data to either gain insights into the dataset or make predictions about future patterns. This model functions as a form of regression analysis, assessing the relationship between one dependent variable and various changing variables. The ARIMA models has the following components: Autoregression (AR), Integrated (I) and Moving average (MA). Autoregression (AR) describes a model where a changing variable is dependent on its own past values or lags. Integrated (I) involves the process of differencing raw observations to achieve stationarity in the time series data, often replacing data values with the difference between current and previous values. Moving average (MA) reflects the relationship between an observation and a residual error, integrating a moving average model with lagged observations.

```python
# Get the train / test data for temperature change and CO2 per capita
temp_train = train.drop("CO2_Per_Capita", axis = 1)
co2_train = train.drop("Temperature_Change", axis = 1)
temp_test = test.drop("CO2_Per_Capita", axis = 1)
co2_test = test.drop("Temperature_Change", axis = 1)

temp_train.reset_index(drop=True, inplace=True)
co2_train.reset_index(drop=True, inplace=True)
temp_test.reset_index(drop=True, inplace=True)
co2_test.reset_index(drop=True, inplace=True)
```

The Autocorrelation Function (ACF) and Partial Autocorrelation Function (PACF) offer valuable insights into time series behaviour. They aid in determining the Autoregressive (AR) and Moving Average (MA) lags for ARIMA models.
